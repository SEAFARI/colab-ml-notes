Seaborn:

Data Visualization Library

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

Note : Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')

#this will be imported in the form of pandas data frame
```

```python
tips.head()
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x ='total_bill',y='tip',col='time',hue='smoker',style='smoker',size='size')
```

```python
# load the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

scatter plot

```python
sns.scatterplot(data=iris,x='sepal_length',y='petal_length',hue='species')
```

```python
# house price dataset

from sklearn.datasets import fetch_california_housing
cali_dataset = fetch_california_housing()

cali_df = pd.DataFrame(cali_dataset.data,columns=cali_dataset.feature_names)
cali_df['PRICE'] = cali_dataset.target
```

```python
print(cali_dataset)
```

```python
cali_df.head()
```

DISTRIBUTION PLOT -

```python
sns.distplot(cali_df['PRICE'])
```

Correlation:

    1. Positive Correlation
    2. Negative Correlation


```python
correlation = cali_df.corr()
```

HEAT MAP

```python
# constructing a Heat Map
plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
